# Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import fetch_20newsgroups
from string import punctuation
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem.snowball import SnowballStemmer
import re
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.cluster import KMeans
from sklearn import metrics
import scipy
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import NMF
from sklearn import preprocessing
```

```python
# fetch data
comall = ['comp.graphics','comp.os.ms-windows.misc','comp.sys.ibm.pc.hardware','comp.sys.mac.hardware']
recall = ['rec.autos','rec.motorcycles','rec.sport.baseball','rec.sport.hockey']
# com_train = fetch_20newsgroups(subset = 'train', categories = comall, shuffle = True, random_state = 42,remove=('headers','footers','quotes'))
# rec_train = fetch_20newsgroups(subset = 'train', categories = recall, shuffle = True, random_state = 42,remove=('headers','footers','quotes'))
# c_train = com_train.data + rec_train.data
com_test = fetch_20newsgroups(subset = 'all', categories = comall, shuffle = True, random_state = 42,remove=('headers','footers','quotes'))
rec_test = fetch_20newsgroups(subset = 'all', categories = recall, shuffle = True, random_state = 42,remove=('headers','footers','quotes'))
c_test = com_test.data + rec_test.data
# label_train = [1] * len(com_train.data) + [0] * len(rec_train.data)
label_test = [0] * len(com_test.data) + [1] * len(rec_test.data)
```

```python
# transform the documents into TF-IDF vectors
stop_words = text.ENGLISH_STOP_WORDS
# stem_words = SnowballStemmer("english")

temp_df3 = c_test
r = re.compile(r'[\s{}]+'.format(re.escape(punctuation)))
for i in range(len(temp_df3)):
    temp_df3[i] = " ".join([t for t in r.split(temp_df3[i])])
    temp_df3[i] = temp_df3[i].replace('\n','').replace('\t','').replace('\r','')
    
vectorizer = TfidfVectorizer(min_df=3, stop_words='english')
tfidf3_test = vectorizer.fit_transform(temp_df3)
print ('The dimensions of the TF-IDF matrix with min_df = 3, exclude the stopwords (no stemming):'+str(tfidf3_test.shape))
```

# Problem 2

```python
# perform K-mean using k = 2
km = KMeans(n_clusters=2, max_iter=100, n_init=1, random_state=42, verbose=0).fit(tfidf3_test)

print("Contingency Table (Confusion Matrix): ")
print(metrics.confusion_matrix(label_test, km.labels_))
print("Homogeneity Score: %0.5f" % metrics.homogeneity_score(label_test, km.labels_))
print("Completeness Score: %0.5f" % metrics.completeness_score(label_test, km.labels_))
print("V-measure: %0.5f" % metrics.v_measure_score(label_test, km.labels_))
print("Adjusted Rand Score: %.5f" % metrics.adjusted_rand_score(label_test, km.labels_))
print("Adjusted Mutual Info Score: %.5f" % metrics.adjusted_mutual_info_score(label_test, km.labels_))
```

# Problem 3

```python
# print('pre')
# singular_values = scipy.linalg.svdvals(tfidf3_test.toarray())
singular_values = scipy.sparse.linalg.svds(tfidf3_test, k=1000, return_singular_vectors=False)
singular_values = np.flip(singular_values, 0)
# print('processing')
ratio_de = tfidf3_test.T.dot(tfidf3_test).diagonal().sum()
ratio_en = np.zeros(1000)
for i in range(1000):
    top_singular = singular_values[i]
    ratio_en[i] = top_singular * top_singular + ratio_en[i-1]
#     print(ratio_en[i])
ratio = ratio_en / ratio_de

plt.figure()
plt.plot(range(1, 1001), ratio*100)
plt.ylabel('percentage')
plt.xlabel('r')
plt.title('the percent of variance the top r principle components can retain v.s. r')
plt.show()
```

```python
rs = [1,2,3,5,10,20,50,100,300]
svd_3 = TruncatedSVD(n_components = 300, n_iter = 10, random_state = 42)
LSI_3 = svd_3.fit_transform(tfidf3_test)
# print(LSI_3)

#Use V-measure to find the best r
vm = []
for r in rs:
    LSI_use = LSI_3[:, :r]
    km = KMeans(n_clusters=2, max_iter=100, n_init=30, random_state=42, verbose=0).fit(LSI_use)
    print('r = %d' %r)
    print("Contingency Table (Confusion Matrix): ")
    print(metrics.confusion_matrix(label_test, km.labels_))
    print("Homogeneity Score: %0.5f" % metrics.homogeneity_score(label_test, km.labels_))
    print("Completeness Score: %0.5f" % metrics.completeness_score(label_test, km.labels_))
    print("V-measure: %0.5f" % metrics.v_measure_score(label_test, km.labels_))
    vm.append(metrics.v_measure_score(label_test, km.labels_))
    print("Adjusted Rand Score: %.5f" % metrics.adjusted_rand_score(label_test, km.labels_))
    print("Adjusted Mutual Info Score: %.5f" % metrics.adjusted_mutual_info_score(label_test, km.labels_))
    print('\n')

#plot V-measure vs r figure
plt.figure()
plt.plot(rs, vm)
plt.ylabel('V-measure Score')
plt.xlabel('r')
plt.title('V-measure score v.s. r (LSI)')
plt.show()
print('Best r is: %d' %rs[np.argmax(vm)])
```

```python
#Use V-measure to find the best r
vm = []
for r in rs:
    nmf_3 = NMF(n_components=r, init='random', random_state=42)
    nmf_use = nmf_3.fit_transform(tfidf3_test)
    km = KMeans(n_clusters=2, max_iter=100, n_init=30, random_state=42, verbose=0).fit(nmf_use)
    print('r = %d' %r)
    print("Contingency Table (Confusion Matrix): ")
    print(metrics.confusion_matrix(label_test, km.labels_))
    print("Homogeneity Score: %0.5f" % metrics.homogeneity_score(label_test, km.labels_))
    print("Completeness Score: %0.5f" % metrics.completeness_score(label_test, km.labels_))
    print("V-measure: %0.5f" % metrics.v_measure_score(label_test, km.labels_))
    vm.append(metrics.v_measure_score(label_test, km.labels_))
    print("Adjusted Rand Score: %.5f" % metrics.adjusted_rand_score(label_test, km.labels_))
    print("Adjusted Mutual Info Score: %.5f" % metrics.adjusted_mutual_info_score(label_test, km.labels_))
    print('\n')
    
#plot V-measure vs r figure
plt.figure()
plt.plot(rs, vm)
plt.ylabel('V-measure Score')
plt.xlabel('r')
plt.title('V-measure score v.s. r (LSI)')
plt.show()
print('Best r is: %d' %rs[np.argmax(vm)])
```

```python
#normalized the data s.t. its variance = 1
B_scaled = B/B.std(axis = 0)
kmenas = km.fit_predict(B_scaled)
plot_kmeans(kmeans, B_scaled, 'K-means clustering, LSI scaled')

A_scaled = A/A.std(axis = 0)
kmeans =km.fit_predict(A_scaled)
plot_kmeans(kmeans, A_scaled, 'K-means clustering, NMF scaled')
```

```python
#log transformation on data after NMF
A_log = np.zeros_like(A)
for j in range(A.shape[0]):
    for k in range(A.shape[1]):
        A_log[j][k] = np.log1p(A[j][k])
kmeans =km.fit_predict(A_log)
plot_kmeans(kmeans, A_log, 'K-means clustering, NMF log transformed')
```

```python
#first, normailze the data first, and then log transformation
A_scaled = A/A.std(axis = 0)
A_log = np.zeros_like(A)
for j in range(A.shape[0]):
    for k in range(A.shape[1]):
        A_log[j][k] = np.log1p(A_scaled[j][k])
kmeans =km.fit_predict(A_log)
plot_kmeans(kmeans, A_log, 'K-means clustering, normalized then log transformed')

#second, log transformation first, then normalize the data
A_log = np.zeros_like(A)
for j in range(A.shape[0]):
    for k in range(A.shape[1]):
        A_log[j][k] = np.log1p(A[j][k])
A_scaled = A_log/A_log.std(axis = 0)       
kmeans =km.fit_predict(A_scaled)
plot_kmeans(kmeans, A_scaled, 'K-means clustering, log transformed then normalized')
```

# problem 5

```python
all_data = fetch_20newsgroups(subset = 'all', shuffle = True, random_state = 42,remove=('headers','footers','quotes'))
labels = all_data.target
# transform the documents into TF-IDF vectors
stop_words = text.ENGLISH_STOP_WORDS
# stem_words = SnowballStemmer("english")

temp = all_data.data
r = re.compile(r'[\s{}]+'.format(re.escape(punctuation)))
for i in range(len(temp)):
    temp[i] = " ".join([t for t in r.split(temp[i])])
    temp[i] = temp[i].replace('\n','').replace('\t','').replace('\r','')
    
vectorizer = TfidfVectorizer(min_df=3, stop_words='english')
tfidf3_test = vectorizer.fit_transform(temp)
print ('The dimensions of the TF-IDF matrix with min_df = 3, exclude the stopwords (no stemming):'+str(tfidf3_test.shape))


def Plot_Contingency_Table(mat, 
                          classes,
                          clusters,
                          title = ''):
    plt.imshow(mat, interpolation = 'nearest', cmap = plt.cm.Blues)
    plt.title(title)
    plt.colorbar()
    plt.xticks(np.arange(len(clusters)), clusters, rotation = 45)
    plt.yticks(np.arange(len(classes)), classes)
    plt.xlabel('Clusters')
    plt.ylabel('Classes')
    plt.show()

class_names = all_data.target_names
cluster_names = []
for i in range(20):
    cluster_names.append("c %s" % str(i+1))

def plot_kmeans_20(data,title):
    color = ["lightcoral", "maroon", "mistyrose", "coral", "peachpuff", "sienna", "orange", "darkgoldenrod",
             "olive", "yellowgreen", "lime", "lightgreen", "g", "seagreen", "tomato", "c", "cadetblue",
             "skyblue", "dodgerblue","grey"]
    cluster_labels = km.labels_
    for i in range(len(labels)):
        plt.scatter(data[i, 0], data[i, 1], color=color[cluster_labels[i]])
    center = km.cluster_centers_
    for i in range(20):
        plt.scatter(center[i, 0], center[i, 1], marker='x', color='k')
    plt.title(title)
    plt.show()

    
    
#lsi
rs = [1,2,3,5,10,20,50,100,300]
svd_3 = TruncatedSVD(n_components = 300, n_iter = 10, random_state = 42)
B = svd_3.fit_transform(tfidf3_test)

vm = []
for r in rs:
    B_used = B[:, :r]
    km = KMeans(n_clusters=20, max_iter=100, n_init=30, random_state=42, verbose=0).fit(B_used)
    print('r = %d' %r)
    print("Homogeneity Score: %0.5f" % metrics.homogeneity_score(labels, km.labels_))
    print("Completeness Score: %0.5f" % metrics.completeness_score(labels, km.labels_))
    print("V-measure: %0.5f" % metrics.v_measure_score(labels, km.labels_))
    vm.append(metrics.v_measure_score(labels, km.labels_))
    print("Adjusted Rand Score: %.5f" % metrics.adjusted_rand_score(labels, km.labels_))
    print("Adjusted Mutual Info Score: %.5f" % metrics.adjusted_mutual_info_score(labels, km.labels_))
    print('\n')

#plot V-measure vs r figure
plt.figure()
plt.plot(rs, vm)
plt.ylabel('V-measure Score')
plt.xlabel('r')
plt.title('V-measure score v.s. r (LSI)')
plt.show()
print('Best r is: %d' %rs[np.argmax(vm)])
```

```python
svd = TruncatedSVD(n_components=50, algorithm='randomized', n_iter=10)
B = svd.fit_transform(tfidf3_test)
B_scaled = B/B.std(axis = 0)

km = KMeans(n_clusters=20, max_iter=100, n_init=1, random_state=42, verbose=0).fit(B_scaled)
print("Contingency Table (Confusion Matrix), using LSI: ")
print(metrics.confusion_matrix(labels, km.labels_))
Plot_Contingency_Table(metrics.confusion_matrix(labels, km.labels_), class_names,cluster_names,
               title = 'Contingency Table after LSI, scaled')

plot_kmeans_20(B_scaled, 'K-means clustering, LSI scaled')
```
